```python
import pandas as pd
```

```python
# Load dataset
df = pd.read_csv("covid_vaccine_statewise.csv")
```

```python
df.describe()
```

```python
print("Columns in dataset:")
print(df.columns)

print("\nNumber of rows:", df.shape[0])
print("Number of columns:", df.shape[1])
```

```python
cols = ['First Dose Administered', 'Second Dose Administered', 'Male (Doses Administered)', 'Female (Doses Administered)']

for col in cols:
    df[col] = pd.to_numeric(df[col], errors='coerce')

# Remove missing values
df = df.fillna(0)
```

```python
first_dose = df.groupby('State')['First Dose Administered'].max().sort_values(ascending=False)

print("\nState-wise First Dose Vaccination:")
print(first_dose)
```

```python
second_dose = df.groupby('State')['Second Dose Administered'].max().sort_values(ascending=False)

print("\nState-wise Second Dose Vaccination:")
print(second_dose)
```

```python
male_total = df['Male (Doses Administered)'].max()

print("\nTotal Males Vaccinated:", int(male_total))
```

```python
import matplotlib.pyplot as plt
```

```python
# Top 10 states by first dose
first_dose.head(10).plot(kind='bar')
plt.title("Top 10 States - First Dose Vaccination")
plt.xlabel("State")
plt.ylabel("Vaccinations")
plt.show()
```
